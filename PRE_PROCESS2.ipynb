### EXRACT FEATURES USING EFFICIENT NET

```python
import torch
import torchvision.transforms as transforms
from torchvision import models
from PIL import Image
```

```python
model = models.efficientnet_b0(pretrained=True)
model.eval()  
```

```python
!pip install torchsummary
from torchsummary import summary
summary(model, (3, 200, 200))
```

```python
# Remove the classification head to use as feature extractor
feature_extractor = torch.nn.Sequential(*(list(model.children())[:-1]))  # Remove classifier
```

```python
# Define image preprocessing pipeline
transform = transforms.Compose([
    transforms.Resize((200, 200)),  # Resize to match model input
    transforms.ToTensor(),  # Convert to tensor
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),  # Normalize
])
```

```python
def processor(image_name):
    image_path = "C:\\Users\\Varshil\\STEEL SURFACE\\Processed_Data\\" + image_name
    image = Image.open(image_path).convert("RGB")
    image = transform(image).unsqueeze(0) # Add batch dimension
    
    # Extract features
    with torch.no_grad():
        features = feature_extractor(image)
    return features
```

```python
filesdf['Features'] = ""
```

```python
filesdf['Features'] = filesdf['filename'].apply(processor)
```

```python
data = filesdf['Features']
```

```python
filesdf['Features'] = data
```

```python
filesdf['Features'] = filesdf['Features'].apply(lambda x: [round(float(x), 5) for x in x.view(1, 1280)[0]])
```
