```python
import pandas as pd
import numpy as np

# necessary imports
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
# models
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
#!pip install xgboost
from xgboost import XGBClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv("ml_ready.csv")
```

```python
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, :-1], df.lables, test_size=0.8, random_state = 69)
```

```python
y_train.value_counts()
```

```python
y_test.value_counts() 
```

- no imbalance in classes -> we can go ahead for machine learning tasks

## LOGISTIC REGRESSION
- since we have 6 classes the objective function will be replaced by softmax.

```python
log = LogisticRegression(max_iter=1000, multi_class='multinomial')
```

```python
log.fit(X_train, y_train)
```

```python
y_pred_log = log.predict(X_test)
```

```python
cf = confusion_matrix(y_test, y_pred_log)
cf
```

### 

### RANDOM FOREST REGRESSOR

```python
rfr = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
rfr.fit(X_train, y_train)
```

```python
y_pred_rfr = rfr.predict(X_test)
```

```python
cf = confusion_matrix(y_test, y_pred_rfr)
cf
```
